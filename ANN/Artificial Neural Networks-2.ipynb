Bu veri setinin doğru çıktısı y={0,1} (Kedi değil, Kedi)

\begin{equation*}
X = 
\begin{bmatrix}
255 \\
231 \\
\vdots \\
x_n
\end{bmatrix}
\end{equation*}

$n_x=64 \times 64 \times 64=12228$ 

$n_x$= Bir veri örneğine ait olan özellik(attribute/feature) sayısı

Verimizi matematiksel olarak ifade edersek;

$(x,y)$ 

$x \in R^{n_x}$, $y \in \{0,1\}$

$m$ adet veri örneği için: $\{(x^{(1)},y^{(1)}),(x^{(2)},y^{(2)}),\dots,(x^{(m)},y^{(m)})\}$

$m=m_{train}$: Eğitim için kullanılacak veri sayısı

$m_{test}$: Test için kullanılacak veri sayısı

\begin{equation*}
X = 
\begin{bmatrix}
\vdots& \vdots&\vdots \\
x^{(1)}&x^{(2)}&x^{(m)} \\
\vdots&\vdots&\vdots \\
\end{bmatrix}
\end{equation*}

\begin{equation*}
Y = 
\begin{bmatrix}
y^{(1)},y^{(1)},\dots,y^{(m)}
\end{bmatrix}
\end{equation*}

$X\in R^{(n_x\times m)}$

$Y \in R^{(1 \times m)}$
